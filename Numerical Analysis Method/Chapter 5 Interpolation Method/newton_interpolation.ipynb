```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ni(x,ans):
    return ans[0]+ ans[1]*(x-0.5)+ ans[2]*(x-0.5)*(x-1)+ ans[3]*(x-0.5)*(x-1)*(x-1.5)+ ans[4]*(x-0.5)*(x-1)*(x-1.5)*(x-2)+ ans[5]*(x-0.5)*(x-1)*(x-1.5)*(x-2)*(x-2.5)
```

```python
X = np.array([0.5,1.0,1.5,2.0,2.5,3.0],dtype=np.float32)
f0 = np.array([1.75,2.45,3.81,4.80,7.00,8.60])
f1 = []
f2 = []
f3 = []
f4 = []
f5 = []
##input = input("Please type in")
```

```python
for i in range(1,6):
    a = (f0[i]-f0[i-1])/(X[i]-X[i-1])
    f1.append(a)
    if i-2 >= 0:
        b = (f1[i-1]-f1[i-2])/(X[i]-X[i-2])
        f2.append(b)
        if i-3 >= 0:
            c = (f2[i-2]-f2[i-3])/(X[i]-X[i-3])
            f3.append(c)
            if i-4 >= 0:
                d = (f3[i-3]-f3[i-4])/(X[i]-X[i-4])
                f4.append(d)
                if i-5 >= 0:
                    e = (f4[i-4]-f4[i-5])/(X[i]-X[i-5])
                    f5.append(e)
                    
    

```

```python
ans = np.array([f0[0],f1[0],f2[0],f3[0],f4[0],f5[0]])
print(ans)
```

```python
output = np.zeros(6)
for i in range(6):
    output[i] = ans[0]+ ans[1]*(X[i]-0.5)+ ans[2]*(X[i]-0.5)*(X[i]-1)+ ans[3]*(X[i]-0.5)*(X[i]-1)*(X[i]-1.5)+ ans[4]*(X[i]-0.5)*(X[i]-1)*(X[i]-1.5)*(X[i]-2)+ ans[5]*(X[i]-0.5)*(X[i]-1)*(X[i]-1.5)*(X[i]-2)*(X[i]-2.5)
print(output-f0)
```

```python
xs = np.linspace(np.min(X),np.max(X),1000,endpoint = True)
ys =[]
for x in xs:
    ys.append(ni(x,ans))
```

```python
plt.title("newton interpolation")
plt.plot(X,f0,'s',label = "original values")
plt.plot(xs,ys,'r',label = "interpolation values")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 4)
plt.show()
```
