```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mis(x,ans):
    return ans[0]+ans[1]*x +ans[2]*(x**2) 
```

```python
X = np.array([0.5,1.0,1.5,2.0,2.5,3.0],dtype=np.float32)
y = np.array([1.75,2.45,3.81,4.80,7.00,8.60])
s = np.ones((6))
x2 = X**2
output = np.zeros(6)
```

```python
A =np.column_stack((s,X,x2))
A_new = np.dot(A.T,A)
y_new = np.dot(A.T,y)
ans = np.dot(np.linalg.inv(A_new),y_new)
```

```python
print(ans)
```

```python
output = np.dot(A,ans)
delta = np.sum((output-y)**2)
print(delta)
```

```python
xs = np.linspace(np.min(X),np.max(X),1000,endpoint = True)
ys =[]
for x in xs:
    ys.append(mis(x,ans))
```

```python
plt.title("interpolation")
plt.plot(X,y,'s',label = "original values")
plt.plot(xs,ys,'r',label = "interpolation values")
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc = 4)
plt.show()
```
