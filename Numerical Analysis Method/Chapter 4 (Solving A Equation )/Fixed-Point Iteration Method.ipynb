```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
x = np.arange(-5,5,0.02)
plt.plot(x,x**3+2*(x**2)+10*x-20,'r--')
plt.axis([0,5,-5,5])
plt.show()
```

```python
x = np.arange(1,2,0.01)
plt.plot(x,20/(x**2+2*x+10),'r--')
plt.axis([1,2,1,2])
plt.show()
```

```python
x = np.arange(1,2,0.01)
plt.plot(x,(-40*x-40)/((x**2+2*x+10)**2),'r--')
plt.axis([1,2,-0.5,0])
plt.show()
```

#As f(x) = x3 + 2 x2 + 10 x - 20, we can obtain that f'(x) = 3 x2 + 4 x + 10 always bigger than 0. So f(x) monotonically increasing when x∈(1,2) 
#x3 + 2 x2 + 10 x - 20 = 0 when x∈(1,2)， no multiple roots

```python
def fo(x):
    return x**3+2*(x**2)+10*x-20

def f(x):
    return 20/(x**2+2*x+10)

def fp(x):
    return (-40*x-40)/((x**2+2*x+10)**2)
```

```python

xlb = 1
xrb = 2
if abs(fp(xlb))<1:
    print("convergent")
else:
    print("divergent")

if 1<f(xrb) and f(xrb)<f(xlb) and f(xlb)<2:
    print("convergent")
else:
    print("divergent")
```

```python
#Iteration
N = 50 
x_itr = 1.5
for i in range(1,N):
    x_itr = f(x_itr)
    if abs(fo(x_itr)) < 0.00001:
        print(i)
        break
print(x_itr)
```
