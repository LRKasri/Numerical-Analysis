```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
x = np.arange(-5,5,0.02)
plt.plot(x,x**3+2*(x**2)+10*x-20,'r--')
plt.axis([0,5,-5,5])
plt.show()
```

```python
def f(x):
    return x**3+2*(x**2)+10*x-20

def fp(x):
    return 3*(x**2)+4*x+10

def f_itr(x):
    return f(x)/fp(x)
```

```python
N = 50 #iteration 
x_itr = 1.5
for i in range(1,N):
    x_itr = x_itr - f_itr(x_itr)
    if abs(f(x_itr)) < 0.00000001:
        print(i)
        break
print(x_itr)
```
